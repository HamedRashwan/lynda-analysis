```python
#Funcation to value_counts two columns
def count_by(col1,col2,v=5):
    """
    col: id x
    col2: catigorical
    """
    return list(df[[col1,col2]].drop_duplicates()[col2].value_counts().head(v).index)
```

```python
count_by('order_id','customer_id')
```

```python
count_by('payment','customer_name')
```

```python
#Funcation to sum two columns
def sum_money(col1,col2,v=5):
    """
    col: id x
    col2: catigorical
    """
    return list(df[[col1,col2]].drop_duplicates().groupby(col1).col2)
```
